```python
#Task 9
word=input()
new=[]
new_1=[]
temp=""
for i in word:
  if i==" ":
    new.append(int(temp))
    temp=""
  else:
    temp+=i
new.append(int(temp))
print(f"Original list: {new}")
for i in new:
  if i%2!=0:
    new_1.append(i)
print(f"Modified list: {new_1}")
```

```python
#Task 10
word="0, 0, 1, 2, 3, 4, 4, 5, 6, 6, 6, 7, 8, 9, 4, 4"
my_list=word.split(", ")
new=[]
for i in my_list:
  new.append(int(i))
print(f"Input list: {new}")
new_1=[]
for i in new:
  if i not in new_1:
    new_1.append(i)
print(f"Modified list: {new_1}")
```
